```python
with open('EXTR_ResBldg.csv') as res_bldg_csv:
    pin_reader = csv.reader(res_bldg_csv)
    next(pin_reader, None)
    pin_zip_dict = {}
    for line in pin_reader:
        pin_zip_dict[line[0]] = line[11]
```

```python
with open('EXTR_RealPropApplHist_V.csv') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    reader = csv.DictReader(csvfile, fieldnames=header)
    zip_prices_dict_2017 = {}
    zip_prices_dict_2012 = {}
    for line in reader:
        if line["RollYr"] == "2017" and line["Major"] in pin_zip_dict:
            if pin_zip_dict[line["Major"]] in zip_prices_dict_2017:
                zip_prices_dict_2017[pin_zip_dict[line["Major"]]].append(int(line["LandVal"]) + int(line["ImpsVal"]))
            else:
                zip_prices_dict_2017[pin_zip_dict[line["Major"]]] = [int(line["LandVal"]) + int(line["ImpsVal"])]
        if line["RollYr"] == "2012" and line["Major"] in pin_zip_dict:
            if pin_zip_dict[line["Major"]] in zip_prices_dict_2012:
                zip_prices_dict_2012[pin_zip_dict[line["Major"]]].append(int(line["LandVal"]) + int(line["ImpsVal"]))
            else:
                zip_prices_dict_2012[pin_zip_dict[line["Major"]]] = [int(line["LandVal"]) + int(line["ImpsVal"])]
        
        
```

```python
import numpy
def median(lst):
    return numpy.median(numpy.array(lst))
```

```python
median_2017_dict = {zipcode: median(zip_prices_dict_2017[zipcode]) for zipcode in zip_prices_dict_2017}
```

```python
median_2012_dict = {zipcode: median(zip_prices_dict_2012[zipcode]) for zipcode in zip_prices_dict_2012}
```

```python
def rate_change(val12, val17):
    return ((val17-val12)/val12)*100
```

```python
apprec_dict = {zipcode: rate_change(median_2012_dict[zipcode], median_2017_dict[zipcode]) for zipcode in median_2017_dict}
```

```python
from operator import itemgetter
```

```python
lst = []
for zipcode in apprec_dict:
    lst.append([zipcode, median_2017_dict[zipcode], apprec_dict[zipcode]])
```

```python
sorted_mdn = sorted(lst, key=itemgetter(1))
```

```python
sorted_apprec = sorted(lst, key=itemgetter(2))
```
